```python
FP = "data/meta/2_clean/dicom.csv"
```

```python
df = pd.read_csv(FP)
stratum = ['image_view', 'abnormality_type', 'cancer', 'assessment']
df[stratum].value_counts()
df.groupby(by=stratum).sample(frac=0.1)
```
