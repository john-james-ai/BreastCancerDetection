# Repository Lab

```python
from bcd.container import BCDContainer
from bcd.dal.file import FileManager
import seaborn as sns
sns.set_style('whitegrid')
```

```python
Config.set_log_level('INFO')
Config.set_mode('exp')
```

```python
container = BCDContainer()
container.init_resources()
container.wire(
    packages=[
        "bcd.dal.repo",  
    ]
)
```

## Task Repository

```python
task_repo = container.dal.task_repo()
meta, tasks = task_repo.get_by_stage(stage_id=0)
meta
```

## Evaluation Repository

```python
eval_repo = container.dal.eval_repo()
evals = eval_repo.get()
evals
```

```python
sns.barplot(data=evals, x='method', y='mse', hue='params')
```

```python
sns.barplot(data=evals, x='method', y='psnr', hue='params')
```

```python
sns.barplot(data=evals, x='method', y='ssim', hue='params')
```

## Reset Repos

```python
uow = container.dal.uow()
#uow.image_repo.delete_by_mode()
uow.task_repo.delete_by_mode()
uow.eval_repo.delete_by_mode()
```
